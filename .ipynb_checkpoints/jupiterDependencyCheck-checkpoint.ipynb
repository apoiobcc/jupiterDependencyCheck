```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from unicodedata import normalize

gradeCurricular = pd.read_html('https://uspdigital.usp.br/jupiterweb/listarGradeCurricular?codcg=45&codcur=45052&codhab=1&tipo=N',
                        match= "Disciplinas");

def pegaRequisitos(codigoDisciplina):
    requisitosDisciplina = pd.read_html("https://uspdigital.usp.br/jupiterweb/listarCursosRequisitos?coddis="+codigoDisciplina);
    return requisitosDisciplina

# oferecimento no formato '2s21' (2° semestre de 2021)
def pegaPassado(oferecimento):
    passadoOferecimento = pd.read_html("https://www.ime.usp.br/~pmiranda/horarios/" + oferecimento + "index.html")
    return passadoOferecimento;

```

## Pegando os códigos das disciplinas sendo oferecidas:

```python
df = gradeCurricular[3]
df.head()
```

```python
codigos = [];
count = 0;
for i in df[0]:
    if(isinstance(i, str)):
        if(re.match("^(([A-Z]{3})[0-9]{4})$", i)):
            codigos.append(i); 
    count +=1;
    # Ao printar os valores de i nota-se que existem valores repetidos após 262,
    # provavelmente um erro de leftovers do jupiter;
    if(count==262):
        break;
print(codigos)
```

## Pegando os requisitos de cada disciplina:

```python
cod = pegaRequisitos(codigos[6]);
req = cod[6];
req.head()
```
